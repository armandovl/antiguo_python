```python
%matplotlib inline 
import pandas as pd #importamos Pandas
#df = pd.read_excel('Datos_Rosa_de_Vientos.xlsx') 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Base_de_Datos= pd.read_excel('Para Ejercicios.xlsx')
```

```python
Base_de_Datos['Grados']
```

```python
Datos=pd.DataFrame(Base_de_Datos)
#lo convertimos en un data frame
```

```python
Datos.groupby('Municipio')['InAI'].mean()
```

```python
Datos.groupby('Municipio')['InAI'].mean().plot(kind='bar')
```

```python
Datos.Grados.groupby(Datos.Municipio).mean().plot(kind='barh')
```

```python
Datos.Grados.groupby(Datos.Municipio).count().plot(kind='pie')
```

```python
Datos.Grados.groupby(Datos.Municipio).mean().plot(kind='box')
```

```python
sns.lmplot(x='Escolaridad',y='Experiencia',data=Datos)
```
