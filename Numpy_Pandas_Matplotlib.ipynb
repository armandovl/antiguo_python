```python
# Resolución de sistemas de ecuaciones

""" supongamos que tenemos el siguiente sistema de ecuaciones y necesitamos resolverlo
X2+X3=5
X1=1
X3=3
"""
#importar numpy como np
import numpy as np

#definir matriz A
A = np.array([[0,1,1],[1,0,0],[0,0,1]])

#definir matriz B
B = np.array([[5],[1],[3]])

#calcular invA
invA = np.linalg.inv(A)


#imprimir Matriz inversa de A
print("Matriz inversa de A es:\n",invA) #para ver cómo se obtiene te coloco el siguiente vídeo del cual saqué el ejercicio
#https://www.youtube.com/watch?v=3BpGef99HEs&t=195s

#calcular X
X = invA.dot(B)


#imprimir Matriz inversa de A
print("X es:\n",X) #para ver cómo se obtiene te coloco el siguiente vídeo para que revisen cómo se multiplican
#las matrices https://www.youtube.com/watch?v=TojmEXYCA2w

print("No subestimemos el poder de numpy")
```

```python
#almacenamos alimentos para distribuir
lista_de_alimentos=['carnes','verduras','bebidas','cereales']
cantidades=[125,423,66,201]
medida=['kg','kg','lt','bulto']

alimentos=pd.Series(cantidades,index=lista_de_alimentos)

print(alimentos)
```

```python
#cuales rubros están pro debajo de 200 para solicitar mercancia
for rubro in alimentos.index:
    if(int(alimentos[rubro])<200):
        print('solicitar {}, debido a que quedan solo {}'.format(rubro,alimentos[rubro]))
```

```python
#si llega nueva mercancia y requiere ser adicionada
lista_de_alimentos=['carnes','verduras','bebidas','cereales']
cantidades=[1000,1000,2000,3000] 
carga_nueva=pd.Series(cantidades,index=lista_de_alimentos)

for rubro in alimentos.index:
    alimentos[rubro]=int(alimentos[rubro])+int(carga_nueva[rubro])

print(alimentos)
```

```python
#Crear elipse
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig1 = plt.figure()
ax1 = fig1.add_subplot(111, aspect='equal')
ax1.add_patch(
    patches.Ellipse(
        (0.5, 0.5),   # centro
        0.8,          # x, long
        0.5,          # y, altura
    )
)

plt.show()
```

```python
#Crear varios círculo
import matplotlib.pyplot as plt
import matplotlib.patches as patches


for i in range(5): #pasar por cada columna
        fig1 = plt.figure()
        ax1 = fig1.add_subplot(111, aspect='equal')
        ax1.add_patch(
            patches.Circle(
                (0.5, 0.5),   # centro
                0.1*(i+1),          # radi0
            )
        )

plt.show()
```

```python
import numpy as np

n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
print(Y1)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va= 'bottom')

for x, y in zip(X, Y2):
    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='center', va= 'top')



plt.show()
```
