```python
%matplotlib inline 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#leer excel
el_excel= pd.read_excel('iris2.xlsx')
```

```python
df=pd.DataFrame(el_excel)
#lo convertimos en un data frame
```

```python
df
```

```python
sns.set(style="ticks")
sns.pairplot(df, hue="Especies")
```

```python

sns.set(style="whitegrid", palette="muted")

# "Melt" the dataset to "long-form" or "tidy" representation
iris1 = pd.melt(df, "Especies", var_name="measurement")

# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="measurement", y="value", hue="Especies",
              palette=["r", "c", "y"], data=iris1)
```

```python
sns.set(style="whitegrid")


# "Melt" the dataset to "long-form" or "tidy" representation
iris4 = pd.melt(df, "Especies", var_name="measurement")

# Initialize the figure
f, ax = plt.subplots()
sns.despine(bottom=True, left=True)

# Show each observation with a scatterplot
sns.stripplot(x="value", y="measurement", hue="Especies",
              data=iris4, dodge=True, jitter=True,
              alpha=.25, zorder=1)

# Show the conditional means
sns.pointplot(x="value", y="measurement", hue="Especies",
              data=iris4, dodge=.532, join=False, palette="dark",
              markers="d", scale=.75, ci=None)

# Improve the legend 
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[3:], labels[3:], title="Especies",
          handletextpad=0, columnspacing=1,
          loc="lower right", ncol=3, frameon=True)
```
