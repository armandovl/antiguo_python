```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
```

```python
#convierte en numerico los valores de la clase 
from sklearn.preprocessing import LabelEncoder

class_le = LabelEncoder()
dataset['variety'] = class_le.fit_transform(dataset['variety'].values)

dataset.to_csv('nuevoiris.csv',index=False)
```

```python
X = dataset.drop('variety', axis=1)  
y = dataset['variety']  
```

```python
from sklearn.model_selection import train_test_split #separa los datos de prueba de los de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50) 
```

```python
from sklearn.tree import DecisionTreeClassifier 

classifier = DecisionTreeClassifier()  
classifier.fit(X_train, y_train) 
```

```python
y_pred = classifier.predict(X_test) 
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_pred))  
print()
print()
print(classification_report(y_test, y_pred)) 
```

```python
accuracy_score(y_test, y_pred)
```
