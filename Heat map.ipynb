```python

%matplotlib inline
import pandas as pd #importamos Pandas
import matplotlib.pyplot as plt #importamos matplotlib
import seaborn as sns #importamos seaborn
Leer_excel= pd.read_excel('heat1.xlsx')
Datos=pd.DataFrame(Leer_excel)#lo convertimos en un data frame
```

```python
Datos
```

```python
flights = Datos.pivot("year", "month", "passengers")

```

```python
flights
```

```python
ax = sns.heatmap(flights)
```

```python
ax = sns.heatmap(flights,annot=True, fmt="d")
```

```python
ax = sns.heatmap(flights,annot=True, fmt="d",cmap="YlGnBu")
```

```python
ax = sns.heatmap(flights,cmap="YlGnBu")
```
