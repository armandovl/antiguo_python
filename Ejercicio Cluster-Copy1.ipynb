```python
%matplotlib inline 
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Base_de_Datos= pd.read_excel('macroeconomia_tarea.xlsx')
Datos=pd.DataFrame(Base_de_Datos)
Datos.head(7)
```

```python
Datos
```

```python
model = KMeans(n_clusters=4)
```

```python
x=Datos.drop('PAIS',axis=1)
```

```python
x.head()
```

```python
X=x.values
```

```python
model.fit(X)
```

```python
centroides= model.cluster_centers_
```

```python
centroides_x=centroides[:,0]
centroides_y=centroides[:,1]
```

```python
plt.scatter(centroides_x, centroides_y, marker='x', s=50)
plt.show()
```

```python
print(model.labels_)
```

```python
etiquetas=model.labels_
Datos_etiquetados=Datos.assign(etiqueta=etiquetas)
mostrar_etiquetas=Datos_etiquetados.drop(["Private_final_consumption","Government_final_consumption.", "Gross_fixed_capital_formation.","Rate_of_growth."], axis=1)
mostrar_etiquetas
```

```python
plt.scatter(X[:,0],X[:,1], c=model.labels_, cmap="rainbow")
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1], c=model.labels_, cmap="rainbow")
plt.scatter(centroides_x, centroides_y, marker='x', s=50, color='black')
plt.show()
```

```python
ks= range(1,10)
inertias=[]

for k in ks:
    model=KMeans(n_clusters=k)
    model.fit(x)
    inertias.append(model.inertia_)
```

```python
plt.plot(ks, inertias, '-D')
plt.xlabel('numero de clusters,K')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```
