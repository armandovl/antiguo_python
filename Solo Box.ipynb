```python
# factorplot en lugar de cat plot
#https://seaborn.pydata.org/whatsnew.html
g = sns.factorplot(x="Genero", y="Edad", capsize=.2, palette="YlGnBu_d", height=6, aspect=.75,
                kind="point", data=Datos)
g.despine(left=True)
```

```python
# factorplot en lugar de cat plot
#https://seaborn.pydata.org/whatsnew.html
g = sns.factorplot(x="Municipio", y="Edad",capsize=.2, palette="YlGnBu_d", height=10, aspect=2.5,
                kind="point", data=Datos)
g.despine(left=True)
```

```python
# factorplot en lugar de cat plot
#https://seaborn.pydata.org/whatsnew.html
g = sns.factorplot(x="Municipio", y="Edad",col="Genero", capsize=.2, palette="YlGnBu_d", height=20, aspect=2,
                kind="point", data=Datos)
g.despine(left=True)
```
